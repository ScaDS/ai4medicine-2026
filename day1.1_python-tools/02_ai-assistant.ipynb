JupyterLab / Jupyter Notebooks offers several options for integrating and using AI assistants that can help us with programming and data analysis tasks.

## API Access

As explained in the [slides](ai-assistant.pdf), we want to use an LLM provider API that our AI assistant can connect to. For this, we need to provide an according API key. To do this for JuypterLab in a virtual environment managed by `uv`:
* Create a file `.env` (hidden file) in the environments' directory
* Add the line: `OPENAI_API_KEY=your-api-key` (replace _your-api-key_ with the actual key)
* Start JupyterLab with the content of this `.env` file via:
```bash
uv run --env-file path/to/.env -- jupyter lab
```

## Import and Initialze the AI Assistant

````python
# Import bia-bob as bob
from bia_bob import bob
# import additional helper functions from bia_bob
from bia_bob import available_models
````

To intialize bob with any LLM provider, we can set the according URL to its API via the parameter `endpoint`:

````python
bob.initialize(endpoint='https://kiara.sc.uni-leipzig.de/api/v1')
````

We can also print all available models for this `endpoint`:

````python
available_models(endpoint='https://kiara.sc.uni-leipzig.de/api/v1')
````

We can now initialize bob with this endpoint and a specific LLM:

````python
bob.initialize(
    endpoint='https://kiara.sc.uni-leipzig.de/api/v1', 
    model="vllm-llama-4-scout-17b-16e-instruct",
)
````

## Use the AI Assistant

The AI assistant can then be used in a code cell via IPython's so-called Magic Commands (`%`, `%%`). 

Either via:

```bash
%bob Single-line prompt. All is written into one line
```

...or via:

```bash
%%bob
Multi-line prompt.
With more details.
And more context.
```


````python
%%bob 
Hello. 
Are you familiar with Machine Learning ? 
Please explain briefly. 
Just a short text output, no code nor code cell !
````

However, a much more useful application of the AI assistant is generating code, e.g. for data analysis tasks. As a general rule, keep in mind that the more specific your prompt is, the better the generated results will be.

**Let's try it out:**

````python
%%bob
Generate Python code to do the following:
- Create a list of 10 random floats
- Compute the mean and the standard deviation
- Print the results
- Make use of the packages numpy and pandas, if necessary
- Put meaningful one-line comments that explain the most important bits
````

````python
# Importing the required package
import numpy as np

# Generating a list of 10 random floats
random_floats = np.random.rand(10)
print(random_floats)

# Calculating mean and standard deviation
mean = np.mean(random_floats)
std_dev = np.std(random_floats)

# Printing mean and standard deviation
print("Mean: ", mean)
print("Standard Deviation: ", std_dev)
````

## System Prompt and Behavior

The basic behavior of an AI assistant (the underlying LLM) can be controlled via a so-called system prompt. A system prompt is a specific instruction to the AI assistant / LLM to control its behavior, tone, or style.

````python
# Import the default bia-bob system prompt
from bia_bob import DEFAULT_SYSTEM_PROMPT
````

````python
print(DEFAULT_SYSTEM_PROMPT[:640])
````

You can configure your own system message:

````python
bob.initialize(
    system_prompt="""
    You are an excellent Python programmer who always 
    puts a smiley comment at the end of the code.
    """
)
````

````python
%bob Generate Python code: a for-loop that prints the numbers from 0 to 10.
````

````python
# Loop through numbers from 0 to 10 and print them
for i in range(11):
    print(i)

# Smiley comment :)
````

You can reset the config to the default system message like this:

````python
bob.initialize(system_prompt=DEFAULT_SYSTEM_PROMPT)
````

You may also define a system prompt for a more tailored behavior fitting your specific domain.  
And we may even ask our AI assistant to do this for us:

````python
%%bob
Write me a short but precise system prompt in markdown format used for an AI assistant in JupyterLab / Notebook.
The goal is to have an assistant for the medical data science domain.
It uses Python for data analysis and preprocessing tasks as well as machine learning and deep learning.
The assistant shall make use of according Python packages
````

````python
medical_system_prompt="""
### System Prompt for Medical Data Science AI Assistant

You are an expert in medical data science and a skilled Python developer. 
Your primary tasks include data analysis, preprocessing, machine learning, and deep learning. 
You use Python with libraries such as numpy, scipy, pandas, matplotlib, seaborn, scikit-learn, and torch. 
You work in a JupyterLab/Notebook environment and utilize ipywidgets for GUI tasks when necessary. 
Your goal is to provide accurate, efficient, and well-documented code solutions.

Key constraints and guidelines:
- Use only the specified libraries: numpy, scipy, pandas, matplotlib, seaborn, scikit-learn, torch, os, ipywidgets.
- Minimize code while ensuring readability and correctness.
- Provide summary, plan, and code sections in markdown format.
- Avoid overwriting predefined variables and functions.

Please respond to user queries with expertise in medical data science and Python development.
"""
````

Great! Let's just use this, shall we:

````python
bob.initialize(system_prompt=medical_system_prompt)
````
